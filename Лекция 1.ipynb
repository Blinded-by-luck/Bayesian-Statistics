Условное распределение на X при условии Y: 
$$ p(x|y) =\frac{p(x, y)}{p(y)}$$

Условное распределение является отношением совместного распределения к безусловному:


$$ Conditional = \frac{Joint}{Marginal} $$

Как распределена величина X при условии, если мы знаем конкретное значение y,
Если величины независимы: $p(x|y) = p(x)$

Правило произведения:
$$p(x, y) = p(x|y)*p(y)$$

Применяя по индукции:
$$ p(x_1, x_2, \dots, x_n) = p(x_n|x_1, \dots, x_{n-1}) * p(x_{n-1}|x_1, \dots, x_{n-2}) \dots p(x_1) $$


Следствие:
$$p(y|x) = \frac{p(x|y)p(y)}{p(x)}$$

$$\int \frac{p(x|y)p(y)}{p(x)} dy = 1$$

Правило суммирования:
$$p(x) = \int p(x|y)p(y)dy =  \int p(x,y)dy $$

$$ \int p(x|y)p(y)dy = E_y p(x|y)$$

$$ p(x_1, \dots, x_k) = \int p(x_1, \dots, x_n) dx_1, \dots, dx_n $$

Формула Байеса:
$$p(y|x) = \frac{p(x|y)p(y)}{\int p(x|y)p(y)dy}$$

$$ Posterior = \frac{Likeliihood * Prior}{Evidence}$$

## Оценка

$$ X = (x_1, \dots, x_n)  \thicksim p_{\Theta}(x) = p(x|\Theta)$$
$\Theta_{ml} = \prod_{i=1}^n p(x_i|\Theta) \rightarrow max$, 
+ Асимптотически нормальны, 
+ Несмещенные, 
+ Оптимальные
+ Дисперсия от истинного значения минимальна 





    Объединение Frequentist и Bayesian
$$ lim_{n \gg d} p(\Theta|x_1, \dots, x_n) = \delta(\Theta - \Theta_{ml})$$
$\delta$ - Дельта функция Дирека, функция вероятность которой сосредоточена в одной точке, 
равенство показывает, что апостериорное распределение параметра переходит к точечной оценке которая равна оценке ммп 

поэтому подход Bayesian - не является отличным от Frequentist, а наоборот приобщает его на достаточно больших n, относительно оценки параметра

## Преимущества Байесовского подхода

    1. Регуляризация (Благодаря не слишком сильной подгонке значения параметра, т.к. мы смотрим на p(x|y) при всех p(y))
    2. Композитность (вложенность вероятности, объединение вероятности при переходе от априорного распределения к апостериорному) 
    3. Обработка данных "на лету" (при доступе к апостериорному распределению дообучить модель можно без доступа к исходным данным)
    4. Построение моделей с латентными переменными
    5. Маштабируемость
